```python
# Número de vértices
N = 10

# Sucesores de cada vértice
grafo = np.zeros((10,10))

# Vecinos de 1
grafo[0,1] = 12; grafo[0,2] = 11; grafo[0,3] = 21

# Vecinos de 2
grafo[1,4] = 21; grafo[1,6] = 16

# Vecinos de 3
grafo[2,1] = 24; grafo[2,5] = 25; grafo[2,7] = 10

# Vecinos de 4
grafo[3,2] = 13; grafo[3,8] = 16

# Vecinos de 5
grafo[4,2] = 7; grafo[4,6] = 9; grafo[4,7] = 11

# Vecinos de 6
grafo[5,3] = 12; grafo[5,7] = 18; grafo[5,8] = 8

# Vecinos de 7
grafo[6,7] = 17; grafo[6,9] = 19

# Vecinos de 8 
grafo[7,9] = 13

#### CAMBIAR A POSITIVO

# Vecinos de 9
grafo[8,7] = 17; grafo[8,9] = 21
```

```python
import networkx as nx
import pylab


#G = nx.DiGraph()
#G.add_edge(1, 2, weight=3)
#G.add_edge(2, 3, weight=5)

G = nx.from_numpy_matrix(grafo, create_using=nx.MultiDiGraph)
G.edges(data=True)

poss = {0: (0, 0), 1: (2, 4), 2: (2,0), 3: (2,-4), 4: (4,2), 5: (4,-2), 6: (6,4), 7: (6,0), 8: (6,-4), 9: (8,0)}

# version 2
#nx.draw(G, pos=nx.spring_layout(G))
# specifiy edge labels explicitly
edge_labels=dict([((u,v,),d['weight'])
for u,v,d in G.edges(data=True)])


nx.draw(G,pos=poss,edge_color='black',width=1,linewidths=1,\
node_size=500,node_color='c',alpha=0.9,\
labels={node:node for node in G.nodes()})

print(edge_labels)
nx.draw_networkx_edge_labels(G,pos=poss,edge_labels=edge_labels)

# show graphs
plt.show()
```
