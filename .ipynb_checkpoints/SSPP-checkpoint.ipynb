# Value iteration 

```python
M = 10
```

```python
def suma(s,a,v0):
    rta = 0
    for j in range(0,M):
        rta += prob(s,j,a)*v0[j]
    return rta
```

```python
parar = False
M = 10
iteracion = 0
v0 = np.zeros(M)
eps = 0.01 # PREGUNTAR
    
start = timer()

while parar == False:
    iteracion += 1
    print( '\n #################### Iteración número ', iteracion, ' ####################\n' )
    vn = np.zeros(M)
    #Itero sobre los estados
    for s in range(0,M-1):        
        As = np.nonzero( grafo[s,:] != 0 )
        As = np.ndarray.tolist(As[0])
        
        valores = np.zeros(len(As))
        
        for i in range(0,len(As)):
            valores[i] = (grafo[s,As[i]] + suma(s,(s,As[i]),v0))
        
        vn[s] = np.min(valores)
        
    #Condición de parada
    if np.linalg.norm( np.subtract( vn,v0 ) ) < eps:
        parar = True

    v0 = vn
    
# Fin del ciclo. Calculo la política
d = np.zeros(M)

for s in range(0,M-1):        
    As = np.nonzero( grafo[s,:] != 0 )
    As = np.ndarray.tolist(As[0])

    valores = np.zeros(len(As))

    for a in range(0,len(As)):
        valores[a] = grafo[s,As[a]] + suma(s,(s,As[a]),v0)

    d[s] = As[np.argmin(valores)]
d[M-1] = M-1
print(d)

end = timer()
print(end - start) 
```

# Policy iteration

```python
def calculo_Pdn_y_rdn(d):
    P = np.identity(M)
    for i in range(0,M):
        P[i,i] = 1
    r = np.zeros(M)
    
    for s in range(0,M-1):
        accion = d[s]
        
        r[s] = grafo[s,int(accion)]
        
        As = np.nonzero( grafo[s,:]!= 0 )
        As = np.ndarray.tolist( As[0] )
        
        for t in As:
            P[s,t] = P[s,t] - prob( s,t,(s,accion) )
    
    return (P,r)
```

```python
start = timer()

d0 = np.zeros(M)
for s in range(0,M-1):
    d0[s] = np.ndarray.tolist(np.nonzero( grafo[s,:] > 0 )[0])[0]
#Condición de parada
parar = False
iteraciones = 1

while( parar == False ):
    print('################# Iteración número: ', iteraciones, '#################\n')
    dn = np.zeros(M)
    
    #Calculamos qué es P_dn y r_dn 
    (P_dn,r_dn) = calculo_Pdn_y_rdn(d0)
                           
    #Quiero resolver el sistema de ecuacones (P_dn)v = r_dn.
    v_n = np.linalg.solve(P_dn,r_dn)
    
    #Cálculo de d_n+1
    #Iteramos sobre todos los estados posibles 
    
    ###########################################
    for s in range(0,M-1):
        #Itero sobre las posibles acciones
        As = np.nonzero( grafo[s,:]!= 0 )
        As = np.ndarray.tolist(As[0])
        
        valores = np.zeros(len(As))
        
        for a in range(0,len(As)):
            valores[a] = grafo[s,As[a]] + suma(s,(s,As[a]),v0)

        valorMinimo = np.min(valores)
        
        if( d0[s] == valorMinimo ):
            dn[s] = d0[s] 
        else:
            dn[s] = As[np.argmin(valores)]
    
    dn[M-1] = M-1
    #Condición de parada:
    parar = (dn==d0).all()
    d0 = dn
    iteraciones += 1
    #Fin del ciclo.
print(dn)
end = timer()
print(end - start)
```

```python
from numpy.random import choice
```

```python
def simulacion():
    nodoActual = 0
    camino = []
    acciones = []
    probabilidades = []
    
    final = False
    while final == False:
        As = np.nonzero( grafo[nodoActual,:]!= 0 )
        As = np.ndarray.tolist(As[0])
        
        accion = np.random.choice(As)
        probs = np.zeros(len(As))
        for a in range(0,len(As)):
            probs[a] = prob(nodoActual, As[a], (nodoActual,accion))
        
        draw = choice(range(0,len(As)), 1, p=probs)
        probabilidades.append(probs[draw[0]])
        camino.append(As[draw[0]])
        acciones.append(accion)
        
        nodoActual = camino[-1]
        
        if nodoActual == M-1:
            final = True
            
    return (camino, probabilidades, acciones)
```
